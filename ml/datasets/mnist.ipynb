```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import tensorflow_datasets as tfds
from train.datasets import get_dataset

%matplotlib inline

data, info = get_dataset('mnist', as_supervised=False)
data_train, data_test = data['train'], data['test']
print(data_train)
print(data_test)
fig = tfds.show_examples(info, data_test)
```

```python
for example in data_train.take(1):  # Only take a single example
    image, label = example["image"], example["label"]
    
    plt.imshow(image.numpy()[:, :, 0].astype(np.float32), cmap=plt.get_cmap("gray"))
    print(f"Label: {label.numpy()}")
```
